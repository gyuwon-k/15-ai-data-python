```python
#주머니에 돈이 있으면 택시를 타고 가고
#주머니에 돈은 없지만 카드가 있으면 택시를 타고 가고
#돈도 없고 카드도 없으면 걸어가라
pocket = ["cash","card"]
if "money" in pocket:
	print("take taxi")
elif "card" in pocket:
	print('take taxi')
else:
	print('walking')	
```

```python
#match-case문
#학점이 A이면 "탁월한 성적입니다"
#학점이 B이면 "우수한 성적입니다"
#C이면 보통입니다
#그외는 노력이 필요합니다
grade = 'b'
if grade == 'a':
	print('탁월한 성적입니다')
elif grade =='b':
	print('우수한 성적입니다')
elif grade =='c':
	print('보통입니다')
else:
	print('노력이 필요합니다')
```

```python
def evaluate_grade(grade):
    match grade.upper(): # 대소문자 구분 없이 처리하기 위해 upper() 사용
        case "A":
            print("탁월한 성적입니다.")
        case "B":
            print("우수한 성적입니다.")
        case "C":
            print("보통입니다.")
        case _: # 위 케이스에 해당하지 않는 모든 경우 (Wildcard)
            print("노력이 필요합니다.")

# 테스트
evaluate_grade("A")
evaluate_grade("B")
evaluate_grade("C")
evaluate_grade("F")
```
