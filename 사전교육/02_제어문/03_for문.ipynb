```python
# for 문의 기본구조
# for 변수 in 리스트(or 튜플, 문자열):
# 	수행할 문장1
# 	수행할 문장2
# 	...
```

```python
test_list = ['one', 'two', 'three']

for li in test_list:
	print(li)
```

```python
#다양한 for 문의 사용
a = [(1,2),(3,4),(5,6)]
for i, j in a:
	print(f"{i} X {j} = {i*j}")
```

```python
marks = [90, 25, 70, 45, 80]
num = 0

for mark in marks:
	num += 1
	if mark >= 60:
		print(f"{num}번 학생은 합격입니다")
	else:
		print(f"{num}번 학생은 불합격입니다")
```

```python
marks = [90, 25, 70, 45, 80]
num = 0

for mark in marks:
	num += 1
	if mark < 60:
		continue
	else:
		print(f"{num}번 학생은 합격입니다")
```

```python
#for 문과 함께 자주 사용하는 range함수
a = range(10)
print(a)
print(list(a))
b = range(1,11)
print(list(b))
```

```python
add = 0
for num in range(1,11):
	add = add + num
print(add)
```

```python
# for와 range 함수를 이용해서 구구단
for i in range(2,10):
	for j in range(1,10):
		print(f"{i} x {j} = {i*j}")
```

```python
for i in range(1,6):
	print('*'*i)
	if i == 5:
		for j in range(6,0,-1):
		 print('*'*j)
```

```python
for i in range(1,6):
	print(f"{' '*(5-i)}{'*'*i}")
	
```

```python
#리스트 컨프리헨션 사용하기
a = [1,2,3,4]
result = []
for num in a:
	result.append(num*3)

print(result)


a = [1,2,3,4]
result = [num*3 for num in a]
print(result)
```

```python
#enumerate 함수 사용하기 - index값을 알려줌
fruits = ["사과", "바나나", "오렌지"]
for number, fruit in enumerate(fruits):
	print(f"{number+1}번째 과일은 {fruit}이다 ")
```

```python
#zip 함수로 여러 리스트 함께 순회하기
names = ["홍길동","유재석","강호동"]
scores = [70, 80, 90]

for i in zip(names,scores):
	print(f"{i[0]}은 {i[1]}점이다")

for (name, score) in zip(names, scores):
	print(f"{name}의 점수는 {score}이다")
```

```python
score = int(input("점수를 입력하시오"))

if not 0<=score<=100:
	print("잘못된 점수입니다")
elif score >= 90:
	print("A")
elif score >= 80:
	print("B")
elif score >= 70:
	print("C")
elif score >= 60:
	print("D")
else:
	print("F")
```
