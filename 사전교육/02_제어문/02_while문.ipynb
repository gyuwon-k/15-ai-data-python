```python
# While 문
# while 조건문:
# 	수행할 문장1
# 	수행할 문장2
#	...
```

```python
#열 번찍어 안넘어가는 나무 없다
tree_hit = 0
while tree_hit < 10:
	tree_hit += 1
	print(f"나무를 {tree_hit}번 찍음")
	if tree_hit == 10:print("나무가 넘어감")
```

```python
# while 문 강제로 빠져나가기
coffee = 5
money = 300
while money:  #0이 아니면 모두 true
	print('돈을 받았으니 커피를 줍니다')
	coffee = coffee -1 
	print(f'남은 커피의 개수는 {coffee}개 입니다')
	if coffee == 0:
		print('커피가 모두 판매되어 판매중지합니다')
		break
	
 	
```

```python
coffee = 5
while True:
	money = int(input("돈을 넣으세요"))

	if money == 300:
		coffee -= 1
		print(f"커피 한 잔드립니다.{coffee}잔 남았습니다")
	elif money >= 300:
		print(f"거스름돈 {money-300}원 드립니다.")
		coffee -= 1
		print(f"커피 한 잔드립니다.{coffee}잔 남았습니다")
	else:
		print(f'돈을 더 넣으세요, {money}원 반환합니다')
	if coffee ==0:
		print("커피가 모두 소진됐습니다. 판매중지합니다")
		break
```

```python
# while문의 처음으로 돌아가기 - continue

a = 0
while a < 10:
	a += 1
	if a%2 ==0:
		continue
	print(a)
```
