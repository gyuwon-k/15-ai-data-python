```python
#딕셔너리
#{key1: value1, key2:value2}
dic = {
	"name" : "ken" ,
	"phone" : "01012341234" ,
	"birth" : "0101"
}
print(dic)
print(type(dic))
```

```python
#딕셔너리에 쌍 추가하기
a = {1: "a"}
a[2] = "b"
print(a)
a["hobby"] = ["running","football"]
print(a)
```

```python
#딕셔너리 요소 삭제하기 (run all하면 오류 풀림)
del a[1]
print(a)
```

```python
#딕셔너리에서 key를 사용하여 value값 얻기
data = {
	"유재석" : "개그맨" ,
	"손흥민" : "축구선수" ,
	"귀도" : "개발자"
}

print(type(data))
print(data["유재석"])
print(data["손흥민"])
data["귀도"] = "코딩"		#있으면 변경
	
data["강호동"] = "개그맨"  	#없으면 추가
print(data)
```

```python
#딕셔너리를 만들 때 주의할 사항
# 1. 딕셔너리에서 key는 고유한 값이므로 중복이 안된다
# 2. 수정이 되는 데이터 타입은 key로 올 수 없다. (리스트)

a = {
	[1,2] : "hi"
}

print(a)
```

```python
a = {
	1: "a",
	"a": 10}

print(a)
```

```python
#딕셔너리 관련 함수
data = {
	"유재석" : "개그맨" ,
	"손흥민" : "축구선수" ,
	"귀도" : "개발자"
}

data.keys()
```

```python
#value 리스트 만들기 -values
data = {
	"유재석" : "개그맨" ,
	"손흥민" : "축구선수" ,
	"귀도" : "개발자"
}

data.values()
```

```python
#key, values 쌍 얻기 -items
data = {
	"유재석" : "개그맨" ,
	"손흥민" : "축구선수" ,
	"귀도" : "개발자"
}
print(list(data.items()))      #리스트로 만듦
```

```python
#key, value 쌍 모두 지우기 -clear
data.clear()
print(data)
```

```python
#key로 value 값 얻기
data = {
	"유재석" : "개그맨" ,
	"손흥민" : "축구선수" ,
	"귀도" : "개발자"
}
print(data.get("유재석"))
print(data.get("김연아"))
print(data("김연아"))                   #get을 쓰는 이유--> 오류 안남
```

```python
#해당 key가 딕셔너리 안에 있는지 조사하기 - in

data = {
	"유재석" : "개그맨" ,
	"손흥민" : "축구선수" ,
	"귀도" : "개발자"
}

print("유재석" in data)
print("김연아" in data)
```

```python
#key로 value 꺼내기 - pop
data = {
	"유재석" : "개그맨" ,
	"손흥민" : "축구선수" ,
	"귀도" : "개발자"
}

data.pop("유재석")        #뽑아내는 것임
print(data)
```
