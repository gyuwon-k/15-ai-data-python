```python
#집합자료형
# 1. 중복을 허용하지 않음
# 2. 순서가 존재하지 않음
# 두 요소의 교집합, 차집합
s1 = set([1,2,3])
s2 = {1,2,3}

print(s1)
print(type(s1))
```

```python
s3 = set("Hello")
print(s3)
```

```python
#교집합 구하기
s1 = set([1,2,3,4,5])
s2 = set([2,4,6,8,10])
print(s1&s2)
s2.intersection(s1)
```

```python
#합집합 구하기
s1 = set(["가","나","다","라","마"])
s2 = set(["마","바","사","아","자"])
print(s1|s2)
print(s2.union(s1))
```

```python
#차집합 구하기
s1 = set(["가","나","다","라","마"])
s2 = set(["마","바","사","아","자"])
print(s1-s2)
print(s2.difference(s1))
```

```python
#집합자료형 관련함수
# 값 1개 추가하기 - add
s1 = set([1,2,3])
s1.add(4)
print(s1)

# 값 여러개 추가하기 -update
s1.update([5,6])
print(s1)

# 특정값 제거하기 - remove
s1.remove(6)
print(s1)

# 모든 값을 제거하기 - clear
s1.clear()
print(s1)
```
