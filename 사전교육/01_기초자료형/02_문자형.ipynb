```python
#문자열연산
##문자열 더해서 연결하기

a = "Python"
b = " is fun"
print(a+b)

##문자열 곱하기

a = "Python"
print(a*3)
```

```python
#문자열 인덱싱
a = "Life is too short, You need Python"
print(a[0])
print(a[5])
print(a[8])
print(a[-1])
```

```python
#문자열 슬라이싱
a = "Life is too short, You need Python"
print(a[0:5])
print(a[8:11])
print(a[28:34])
print(a[ :4])
print(a[-6: ])
```

```python
#문자형 포메팅 따라하기
# 1. 숫자 바로 대입
"I eat %d apples" % 3

# 2. 문자열 바로 대입
print("I eat %s apples" % "five")

# 3. 숫자값을 나타내는 변수로 대입
num = 3
print("I eat %d apples" %num)

# 4. 두 개 이상의 값 넣기
num = 10
day = "Three"
print("I ate %d apples, so I was sick for %s days" %(num,day))
```

```python
# 문자열 포맷 코드
# %s - 문자열
# %d - 정수
# %f - 실수
# %o - 8진수
# %x - 16진수
# %c - 문자 1개
```

```python
#정렬과 공백
print("%10s" % "hi")
```

```python
#소수점 표현하기
print("%.4f" % 3.4213234)
```

```python
#숫자 바로 대입하기
print("I eat {0} apples".format(3))

#문자열 바로 대입하기
print("I eat {0} apples".format("five"))

#두 개 이상의 값 넣기
num = 10
day = "three"
print("I ate {0} apples, so I was sick for {1} days".format(num,day))
```

```python
#f문자열 포매팅
name = "홍길동"
age = 30
print(f"나의 이름은 {name} 나이는 {age}살 입니다")
```

```python
#문자열 관련 함수들
# 1. 문자 개수 세기 - count
a = "hobby"
a.count("b")
```

```python
# 2. 위치 알려주기 함수 - find 찾는게 없더라도 에러가 발생하지 않도록 할 때 사용(+list에서 사용불가)
a = "Python is the best choice"
a.find("b")
```

```python
# 3. 위치 알려주기 함수2 -index 무조건 찾아야할 때만 사용(없을경우 error발생)
a = "Life is too short"
a.index("t")
```

```python
# 4. 문자열 삽입 - join
",".join("abcd")
```

```python
# 5. 소문자를 대문자로 - upper
a = "hi"
a.upper()
```

```python
# 6. 대문자를 소문자로 - lower
a = "HI"
a.lower()
```

```python
# 7. 왼쪽 공백지우기 -lstrip
a = "    Life"
a.lstrip()
```

```python
# 8. 오른쪽 공백지우기 - rstrip
a = "Life    "
a.rstrip()
```

```python
# 9. 양쪽 공백지우기 -strip
a = "   hi   "
a.strip()
```

```python
# 10. 문자열 바꾸기 - replace
a = "Life is too short"
a.replace("Life","Leg")
```
