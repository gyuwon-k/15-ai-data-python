```python
# 리스트는 어떻게 만들고 사용할까?
a = []  # a= list()
b = [1,2,3]
c = ["life", "is", "too", "short"]
d = [1, 2, "life", "is"]
e = [1, 2, ["life", "is"]]

#a = int("1") 숫자를 문자로 바꿈
```

```python
# 리스트의 인덱싱
a = [1, 2, 3] #print(type(a)) -->list
print(a[0])
print(a[2])
a = [1, 2, 3, ["a", "b", ["life", "is"]]]
print(len(a))
print(a[-1])
print(a[-1][1])
print(a[-1][2][0])
```

```python
#f리스트의 슬라이싱
a = [1, 2, 3, 4, 5]
print(a[0:3])
print(a[:5])
print(a[0:])
print(a[1:3])

a = [1,2,3,["a","b","c"],4, 5]
print(a[3][:2])
```

```python
# 리스트 연산하기
a = [1,2,3]
b = [4,5,6]
print(a+b)
print(a*3)
```

```python
# 리스트의 수정과 삭제
a = [1, 2, 3]
b = [4,5,6]
a[2] = 4
print(a)

del a[1]
print(a)

del b[:2]
print(b)
```

```python
# 리스트관련 함수
# 1. 리스트에 요소 추가하기 -append
## []가 포함되지 않은 채 리스트 삽입하기 위해선 extend사용
a = [1,2,3]
a.append(4)
print(a)

a.append([5,6])
print(a)
```

```python
# 2. 리스트 정렬 -sort
a = [1, 4, 3, 5, 2]
b = [1, 4, 3, 5, 2]
a.sort()
print(a)

b.sort(reverse=True)
print(b)
```

```python
# 3. 리스트 뒤집기 - reverse
a = [1, 4, 3, 5, 2]
a.reverse()
print(a)
```

```python
# 4. 인덱스 반환
a = [1,2,3]
a.index(3)
```

```python
# 5. 리스트에 요소 삽입 - insert
a = [1,2,3]
a.insert(3,4)
print(a)
```

```python
# 6. 리스트 요소 제거 - remove
a = [1,2,3,1,2,3]
a.remove(3)
print(a)
```

```python
# 7. 리스트 요소 제거하기 - pop
a = [1,2,3]
a.pop()
a.pop()
print(a)
```

```python
# 8. 리스트에 포함된 요소 x의 개수 세기 - count
a = [1,2,3,1]
a.count(1)
```

```python
# 9. 리스트 확장 - extend
a = [1,2,3]
a.extend([4,5])
print(a)
```
