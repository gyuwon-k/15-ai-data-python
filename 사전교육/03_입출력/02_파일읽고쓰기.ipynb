```python
# 파일 생성하기
f = open("./새파일.txt", "w")
f.close()
```

```python
f = open("./새파일2.txt", "w")

for i in range(1,11):
	data = f"{i}번째 줄입니다\n"
	f.write(data)
f.close()
```

```python
f = open("새파일2.txt","r")
while True:
	line = f.readline()
	print(line)
	if not line:
		break
f.close()
```

```python
#readlines 함수 사용하기
f = open("새파일2.txt","r")
lines = f.readlines()
for line in lines:
	print(line)
f.close()
```

```python
# read() 함수 사용하기
f = open("./새파일2.txt","r")
data = f.read()
print(data)
f.close()
```

```python
# 파일에 새로운 내용 추가하기
f = open("새파일2.txt","a")

for i in range(11,21):
	data = f"{i}번째 줄입니다\n"
	f.write(data)
f.close()
```

```python
# with문과 함께 사용하기

with open("새파일3.txt","w") as f:
	f.write("Life is too short, You need Python")
```
