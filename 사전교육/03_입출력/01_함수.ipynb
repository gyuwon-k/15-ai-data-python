```python
# # 함수

# def 함수이름(매개변수1, 매개변수2...):
# 	수행할 문장1
# 	수행할 문장2
# 	...

def add(a, b):
	return a+b
```

```python
# 1. 일반적인 함수
# 입력값 O, 반환값 O
def add(a, b):
	result = a+b
	return result
c = add(3,4)
print(c)
```

```python
# 2. 입력값이 없는 함수
# 입력값 X, 반환값 O
def say():
	return "hi"
c = say()
print(c)
```

```python
# 3. 반환값이 없는 함수
# 입력값 O, 반환값 X
def add(a,b):
	print(a+b)
c = add(3,4)
print(c)
```

```python
# 4. 입력값, 반환값이 없는 함수
# 입력값 X, 반환값 X
def say():
	print("hi")
c = say()
print(c)
```

```python
# 매개변수를 지정하여 호출하기
def sub(a,b):
	return a - b
c = sub(b=3, a=7)
print(c)
c = sub(a=7, b=3)
print(c)
c = sub(7,3)
print(c)
c = sub(3,7)
print(c)
```

```python
# 입력값이 몇 개가 될지 모를 때
def add_many(*args):
	result = 0
	print(args,type(args))
	for i in args:
		result += i
	return result
c = add_many(1,2,3,4,5,6)
print(c)
```

```python
def order_coffee(menu, size="Tall", shot=1):
	print(f"[주문] {menu} ({size}, {shot}샷)")

order_coffee("아메리카노")
order_coffee("라떼", "Grande") 
order_coffee("에스프레소", "Short", 3)
```

```python
#키워드 매개변수 kwargs
def print_kwargs(**kwargs):
	print(kwargs)

print_kwargs(a=1)
print_kwargs(name = "홍길동", age = 20)
print_kwargs(name = "kim", age = 25, city = "대구")
```

```python
def creat_profile(**profile):
	print("="*5, "프로필정보", "="*5)
	for key, value in profile.items():
		print(f"{key} : {value}")

creat_profile(name = "kim", age = 25, city = "대구", hobby = "러닝")

```

```python
# 매개변수의 초기값 미리 설정하기
def say_myself(name, age, man=True):
	print(f"나의 이름은 {name}입니다.\n나의 나이는 {age}살 입니다.")
	if man == True:
		print("남자입니다")
	else:
		print("여자입니다")

say_myself("홍길동",27)
say_myself("김연아",30,False)
```

```python
#함수 안에서 선언한 변수의 효력범위
#함수 안의 변수와 밖의 변수가 같은 문자라도 다르게 작용함
a = 1
def vertext(a):
	a += 1

vertext(a)
print(a)
```

```python
#global 잘안씀
a =1 
def vertext():
	global a
	a += 1
vertext()
print(a)
```

```python
#lambda 예약어
# 함수_이름 = lambda 매개변수1, 매개변수2, ...: 수행할 문장1

# def add(a,b):
# 	return a+b


add = lambda a,b : a+b
print(add(3,3))
```

```python
# 함수의 독스트링(Docstring)

def add(a,b):
	""" 
	두 숫자를 더하는 함수

	parameter
	a(int,float) : 첫 번째 숫자
	b(int,float) : 두 번째 숫자

	return
	int, float: 두 숫자의 합
	"""
	return a+b
print(add.__doc__)
```
