```python
## 막대그래프
##판드스객체.plot(kind ="bar")

import pandas as pd

df = pd.read_excel("./data/남북한발전전력량.xlsx")

df_ns = df.iloc[[0,5],3:]
df_ns.index = ['South', 'North']
df_ns = df_ns.astype(int)

tdf_ns = df_ns.T
print(tdf_ns.head())
print()
tdf_ns.plot(kind="bar");
```

```python
## 수평 막대 그래프
tdf_ns.plot(kind="barh");
```

```python
## 누적 막대 그래프
tdf_ns.plot(kind="bar", stacked=True);
```

```python
## 히스토그램
import pandas as pd

df = pd.read_excel("./data/남북한발전전력량.xlsx")

df_ns = df.iloc[[0,5],3:]
df_ns.index = ['South', 'North']
df_ns = df_ns.astype(int)

tdf_ns = df_ns.T
print(tdf_ns.head())
print()
tdf_ns.plot(kind="hist");
```

```python
tdf_ns.plot(kind='hist',
			alpha = 0.5,
			figsize= (6,2),
			grid=True,
			color = 'skyblue',
			edgecolor = 'black');
```

```python
tdf_ns.hist(figsize=(6,2));
```
