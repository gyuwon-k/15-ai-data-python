```python
# data.go.kr
# 국토교통부 아파트 매매 실거래가 자료

import requests
import xml.etree.ElementTree as ET
import pandas as pd

service_key = "secret"

url = "https://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade"

def fetch_apt_trade(servicekey, lawd_cd, deal_ymd):
    params = {
        "serviceKey": servicekey,
        "LAWD_CD": lawd_cd,
        "DEAL_YMD": deal_ymd,
        "numOfRows": 10,
        "pageNo": 1
    }

    response = requests.get(url, params=params)
    root = ET.fromstring(response.content)
    items = root.findall(".//item")
    records =[]
    
    for item in items:
        row={}
        for child in item:
            row[child.tag] = child.text
        records.append(row)
    return records

data = fetch_apt_trade(service_key, lawd_cd="11110", deal_ymd="202604")
df = pd.DataFrame(data)
df.head()
```

```python
import requests

service_key ="secret"
url = "https://apis.data.go.kr/B552696/countrygrade/credit-grade"

def country_credit_ration(page_no=1, num_of_row = 100):

	params = {
		"serviceKey": service_key,
		"pageNo":page_no,
		"numOfRows":num_of_row
	}
	response = requests.get(url, params=params)

	results = response.json()
	print(results)

country_credit_ration(1,100)
```
