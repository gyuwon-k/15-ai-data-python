```python
# 행 인덱스 재배열
dict_data = {
	"c0" : [1,2,3],
	"c1" : [4,5,6],
	"c2" : [7,8,9],
	"c3" : [10,11,12],
	"c4" : [13,14,15]
}

df = pd.DataFrame(dict_data, index = ['r0','r1','r2'])
df
```

```python
# 행 인덱스 초기화

import pandas as pd

dict_data = {
	"c0" : [1,2,3],
	"c1" : [4,5,6],
	"c2" : [7,8,9],
	"c3" : [10,11,12],
	"c4" : [13,14,15]
}

df = pd.DataFrame(dict_data, index = ['r0', 'r1', 'r2'])
print(df)
print()

ndf = df.reset_index()
print(ndf)
print()

ndf2 = df.reset_index(names=['C00'])
print(ndf2)
print()

#***
ndf3 = df.reset_index(drop=True)
print(ndf3)
```

```python
# 행 인덱스를 기준으로 데이터프레임 정렬

import pandas as pd

dict_data = {
	"c0" : [1,2,3],
	"c1" : [4,5,6],
	"c2" : [7,8,9],
	"c3" : [10,11,12],
	"c4" : [13,14,15]
}

df = pd.DataFrame(dict_data, index = ['r0', 'r1', 'r2'])

ndf = df.sort_index(ascending=False)
print(ndf)
print()
# print()
# print()


ndf2 = df.sort_index(ascending=True)
ndf2
```

```python
# 열 기준 정렬

import pandas as pd

dict_data = {
	"c0" : [1,2,3],
	"c1" : [4,5,6],
	"c2" : [7,8,9],
	"c3" : [10,10,11],
	"c4" : [15,14,14]
}

df = pd.DataFrame(dict_data, index = ['r0', 'r1', 'r2'])

ndf = df.sort_values(by='c1', ascending= False)
print(ndf)
print()

ndf2 = df.sort_values(by='c2', ascending=True)
print(ndf2)
print()

# 중복된 숫자일 경우 먼저 나온 것을 기준으로 하고 정렬함
ndf3 = df.sort_values(by=['c3','c4'], ascending=[False, True])
ndf3
```
