```python
# 시리즈
## 외부라이브러리 pandas설치

##라이브러리 불러오기
import pandas as pd

dict_data = {
	'a' : 1,
	'b' : 2,
	'c' : 3
}

sr = pd.Series(dict_data)
print(type(sr))
print(sr)
```

```python
## 인덱스
## series객체.index
print(sr.index)
```

```python
## 데이터 값 배열
## Series객체.values

print(sr.values)
```

```python
## 시리즈 인덱스
import pandas as pd
list_data = ['2019-01-02', 3.14, 'ABC', 100, True]
sr = pd.Series(list_data)
print(sr)
```

```python
## 인덱스
print(sr.index)

## values
val = sr.values
print(val)

## 시리즈 객체의 데이터 타입 확인
print(sr.dtype)

## 시리즈 객체의 원소 개수 확인
print(len(sr))

## 시리즈 배열의 형태 확인 / shape 속성
print(sr.shape)

## 데이터 구조의 차원 수
print(sr.ndim)
```

```python
# 원소선택
import pandas as pd

tup_data = ('영인', '2010-05-01','여',True)
sr = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부'])
print(sr)



print(sr.iloc[0])
# print(sr[0])   #인덱스가 문자열인 경우만 가능.
print()
print(sr['이름'])


# print(sr[1,2])
print('\n')
print(sr[['생년월일', '성별']])

print(pd.Series(5))
print()
print(pd.Series(5, index = ['a','b','c']))
```
