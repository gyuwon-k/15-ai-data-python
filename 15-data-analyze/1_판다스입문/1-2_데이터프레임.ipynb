```python
# 행 인덱스 / 열 이름 설정
import pandas as pd
df = pd.DataFrame([[15,'남','덕영중'],[17,'여','수리중']],
				   index = ['준서','예은'],
				   columns=['나이','성별','학교'])
print(df)
```

```python
# rename을 사용하여 인덱스/칼럼 명 변경    **재할당하지않으면 안바뀜
import pandas as pd
df = pd.DataFrame([[15,'남','덕영중'],[17,'여','수리중']],
				   index = ['준서','예은'],
				   columns=['나이','성별','학교'])

df = df.rename(columns={'나이' : '연령', '성별' : '남녀', '학교':'소속'})
print(df)
```

```python
# 행/열 삭제
## 행 삭제

import pandas as pd

exam_data = {'수학' :[90,80,70], '영어':[98, 89, 95],
			 '음악' :[85,95,100], '체육' :[100,90,90]}

df = pd.DataFrame(exam_data, index=['서준','우현','인아'])
print(df)

df2 = df.copy()
df2 = df2.drop('우현')         # 지우고 재할당 꼭 해야함
print(df2)
```

```python
#axis는 행을 삭제할 것인지 열을 삭제할 것인지 선택

df3 = df.copy()
df3 = df3.drop(['우현', '인아'], axis=0)
print(df3)
print()

df4 = df.copy()
df4 = df4.drop(['우현', '인아'], axis='index')
print(df4)
print()

df5 = df.copy()
df5 = df5.drop(index = ['우현'])
print(df5)
```

```python
## 열 삭제
import pandas as pd

exam_data = {'수학' :[90,80,70], '영어':[98, 89, 95],
			 '음악' :[85,95,100], '체육' :[100,90,90]}

df = pd.DataFrame(exam_data, index = ['서준', '우현', '인아'])

df2 = df.copy()
df2 = df2.drop('수학', axis=1)
print(df2)
print()

df3 = df.copy()
df3 = df3.drop(['영어', '음악'],axis =1)
print(df3)
print()

df4 = df.copy()
df4 = df4.drop(['영어', '음악'],axis = 'columns')
print(df4)
print()

df5 = df.copy()
df5 = df5.drop(columns=['수학'])
print(df5)
```

```python
##원소 값 변경
import pandas as pd

exam_data = {'이름' : ['서준','우현','인아'],
			'수학' :[90,80,70], '영어':[98, 89, 95],
			 '음악' :[85,95,100], '체육' :[100,90,90]}

df = pd.DataFrame(exam_data)

df = df.set_index('이름')

df.iloc[0,3] = 80
# df.loc['서준']['체육'] = 90        ---->적용안됨/ 대괄호연속사용x 
df.loc['서준','체육'] = 100
df
```

```python
# 행, 열의 위치 바꾸기

import pandas as pd

exam_data = {'이름' : ['서준','우현','인아'],
			'수학' :[90,80,70], '영어':[98, 89, 95],
			 '음악' :[85,95,100], '체육' :[100,90,90]}
df = pd.DataFrame(exam_data)

df = df.transpose()
print(df)
print()

df = df.T
print(df)
```
