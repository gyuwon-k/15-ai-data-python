```python
#시리즈 vs. 시리즈
# Series1 + 연산자(+,-,*,/) + Series2
import pandas as pd

student1 = pd.Series({'국어' : 100,
					  '영어' : 80,
					  '수학' : 90})
student2 = pd.Series({'국어' : 80,
					  '영어' : 90,
					  '수학' : 80})

addition = student1 + student2
print(addition)
print(type(addition))
print()

subtraction = student1 - student2
print(subtraction)
print(type(subtraction))
print()

mul = student1 * student2
print(mul)
print(type(mul))
print()

div = student1 / student2
print(type(div))
print(div)

result = pd.DataFrame([addition, subtraction, mul,div], index=['덧셈', '뺄셈', '곱셈','나눗셈'])
result
```

```python
# NaN 값이 있는 시리즈 연산
import pandas as pd
import numpy as np

student1 = pd.Series({'국어' : np.nan,
					  '영어' : 80,
					  '수학' : 90})
student2 = pd.Series({'수학' : 80,
					  '국어' : 90})

add = student1 + student2
sub = student1 - student2
mul = student2 * student1
div = student1 / student2

print(type(div))
print()

result = pd.DataFrame([add, sub, mul,div], index=['덧셈', '뺄셈', '곱셈','나눗셈'])
result
```

```python
# 연산 메소드
import pandas as pd
import numpy as np

student1 = pd.Series({'국어' : np.nan,
					  '영어' : 80,
					  '수학' : 90})
student2 = pd.Series({'수학' : 80,
					  '국어' : 90})

sr_add = student1.add(student2, fill_value=0)
sr_sub = student1.sub(student2, fill_value=0)
sr_mul = student1.mul(student2, fill_value=0)
sr_div = student1.div(student2, fill_value=0)

result = pd.DataFrame([sr_add,sr_sub,sr_mul,sr_div], index=['덧셈', '뺄셈', '곱셈','나눗셈'])
result
```

```python
# 데이터프레임끼리 더하기

import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]

add = df + 10

sub = add - df
# 끝에서 5개만 출력
print(sub.tail())
print(type(sub))
```

```python
sample1 = add.tail(10).fillna(0.0)
sample1.shape
sample1

```

```python
sample2 = sub.tail(10).fillna(0)
sample2.shape
sample2
```

```python
df_add = sample1.add(sample2, fill_value=0)
df_sub = sample1.sub(sample2, fill_value=0)
df_mul = sample1.mul(sample2, fill_value=0)
df_div = sample1.div(sample2, fill_value=0)

print(df_add)
print()
print(df_sub)
print()
print(df_mul)
print()
print(df_div)
```
