```python
## 문자열 메서드

import pandas as pd
import numpy as np

ser = pd.Series(["Apple_사과","Banna_바나나","Cherry_체리",np.nan])

serr = pd.Series(ser.values, index=["First ", "Second ","Third ", "Fourth"], dtype= "string")
serr
```

```python
# 소문자 변환
print(serr.str.lower())
print()
# 대문자 변환
print(serr.str.upper())
print()
#문자열 길이 계산
print(serr.str.len())
```

```python
# 문자열 분할
print(serr.str.split("_"))
print()

# 문자열 분할하여 DataFrame으로 확장
print(serr.str.split("_", expand=True))
print()

# 분할된 문자열 중 특정 요소 선택
print(serr.str.split("_").str.get(1))
```

```python
# 행 인덱스 추출
idx = serr.index
print(idx)

# 양쪽 공백 제거
print(idx.str.strip())

#왼쪽 공백 제거
print(idx.str.lstrip())

#오른쪽 공백 제거
print(idx.str.rstrip())
```

```python
# 문자열 교체
print(serr.str.replace("_",":", regex = False))
print()

#정규표현식을 사용한 문자열 교체
print(serr.str.replace("[^a-zA-Z\s]","",regex=True))
```

```python
## 문자열 슬라이싱

print(serr.str[0])
print()

print(serr.str[0:4])
```

```python
# 특정 문자 포함 여부 확인
contains_A = serr.str.contains("A", na=False)
print(contains_A)
print()

#패턴 포함여부확인
contains_pattern = serr.str.contains(r"[A|B][a-z]+")
print(contains_pattern)
print()

# 전체 문자열의 패턴 일치 여부 확인
fullmathcch_pattern = serr.str.fullmatch(r"[A|B][a-z]+")
print(fullmathcch_pattern)
```

```python
list_data = ["a1","b2","c3","d4a5","e6e7","f8"]
ser3 = pd.Series(list_data,dtype= "string")

digits = ser3.str.extractall(r"(\d)").groupby(level=0).agg(''.join)
print(digits)

letters = ser3.str.extractall(r"([a-z])").groupby(level=0).agg(''.join)
print(letters)
```
