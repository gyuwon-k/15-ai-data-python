#세트 사용하기

```python
#세트 만들기
## 세트 = {값1, 값2, 값3}
## 1. 순서가 정해져 있지 않다.
## 2. 중복될 수 없다.
fruits = {'strawberry', 'grape', 'orange', 'pineapple', 'cherry'}
fruits
```

```python
# 세트에 특정값이 있는지 확인
fruits = {'strawberry', 'grape', 'orange', 'pineapple', 'cherry'}
print("orange" in fruits)
print("peach" in fruits)
```

```python
#set함수를 사용하여 만들기
## set(반복가능한 개체)
a = set('apple')
print(a)
b = set(range(5))
print(b)
c = set()
print(c)
print(type(c))

#비어있을 땐 dict형태
d = {}
print(d)
print(type(d))
```

```python
#집합 연산 사용
## 세트1 | 세트2
## set.union(세트1, 세트2)

a = {1,2,3,4}
b = {3,4,5,6}
print(a|b)
print(set.union(a,b))
print(a.union(b))
```

```python
# 교집합
print(a&b)
print(set.intersection(a,b))
print(a.intersection(b))
print(b.intersection(a))
```

```python
#차집합
print(a-b)
print(set.difference(a,b))
print(a.difference(b))
print('='*10)
print(b-a)
print(set.difference(b,a))
print(b.difference(a))
```

```python
#대칭자집합
print(a^b)
print(set.symmetric_difference(a,b))
```

```python
#집합 연산 후 할당 연산자 사용하기

## 세트 1 |= 세트 2
## 세트1.update(세트2)
a = {1,2,3,4}
a |= {5}
print(a)
a.update({6})
print(a)
```

```python
## 세트1 &= 세트 2
## 세트1.intersection_update(세트2)

a = {1,2,3,4}
a &= {0,1,2,3,4}
print(a)
a.intersection_update({3,4}) #update를 해줌으로써 할당됨 / intersection만 사용시 다른 문자에 할당해주고 출력해야함\
print(a)
```

```python
## 세트1 ^= 세트2
## 세트1.symmetric_difference_update(세트2)

a = {1,2,3,4}
a ^= {3,4,5,6}
print(a)

a = {1,2,3,4}
a.symmetric_difference_update({3,4,5,6})
print(a)
```

```python
# 부분 집합과 상위집합 확인하기
## 현재 세트 <= 다른세트
## 현재세트.issubset(다른세트) 현재세트가 다른 세트의 부분집합인지 확인

a = {1,2,3,4}
print(a<={1,2,3,4})
print()
print(a.issubset({1,2,3,4,5}))
print()

## 현재세트 < 다른세트
print(a<{1,2,3,4,5})
print()

## 현재세트 >= 다른세트
## 현재세트.issuperset(다른세트)
print(a >= {1,2,3,4})
print()
print(a.issuperset({1,2,3,4}))
print()

## 현재세트 > 다른세트
print(a > {1, 2, 3})
print()

## 현재세트 = 다른세트 확인
print(a == {1, 2, 3, 4})
print()
print(a == {4, 2, 1, 3})
```

```python
# 세트가 겹치지 않는지 확인하기
## 현재세트.isdisjoint(다른세트)
## 겹치는 요소가 없으면 True, 있으면 False

a = {1,2,3,4}
print(a.isdisjoint({5, 6, 7, 8}))
print(a.isdisjoint({3, 4, 5, 6}))
```

```python
# 세트에 요소 추가하기
a = {1, 2, 3, 4}
a.add(5)
print(a)
# print(a.add(5)) --> None값 발생 / 수정만 하고 값을 반환하지 않기에
```

```python
# 세트에 특정 요소 삭제하기
a = {1, 2, 3, 4}
a.remove(4)
print(a)
# a.remove(10)  ---> key error발생
# print(a.remove(3)) --> None값 발생
a.discard(2)
print(a)

a.discard(10)  # --> remove와 달리 error가 발생하지않음
print(a)
```

```python
# 세트에서 임의의 요소를 삭제하기
a = {1, 2, 3, 4}
a.pop() 	#-->임의의 요소가 삭제됨 / 마지막 요소가 아님 / 소괄호에 아무 인수도 못들어감
print(a)
```

```python
# 세트의 할당과 복사
a = {1, 2, 3, 4}
b = a
print(a is b)
b.add(50)
print(a)
```

```python
a = {1, 2, 3, 4}
b = a.copy()
b.add(5)
print(a is b)
print(a==b)
```

```python
#반복문으로 세트의 요소 모두 출력하기
a = {1, 2, 3, 4}
for i in a:
	print(i)
```

```python
# 세트 컴프리헨션
a = {i for i in 'apple'}
print(a)
b = {i for i in range(5)}
print(b)

# 세트 표현식에 if조건문 사용하기
## 1. {식 for 변수 in 세트 if 조건문}
## 2. set(식 for 변수 in 세트 if 조건문)
c = {i for i in 'pineapple' if i not in 'apl'}
print(c)
```
