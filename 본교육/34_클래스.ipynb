```python
#클래스와 매서드 만들기

# class 클래스이름:
# 	def 매서드(self):
# 		코드

class Person:
	def greeting(self):
		print('Hello')

# 인스턴스 = 클래스
james = Person()
james.greeting()
```

```python
# 속성 사용하기
# class 클래스이름:
#	 def __init__(self):
			# self.속성 = 값

class Person:
	def __init__(self):
		self.hello = "안녕하세요"
	
	def greeting(self):
		print('hello')
```

```python
# # 인스턴스를 만들 때 값 받기
# class 클래스 이름:
# 	def __init__(self, 매개변수1, 매개변수2):
# 		self.속성1 = 매개변수1
# 		self.속성2 = 매개변수2

class Person:
	def __init__(self,name, age, address,wallet):
		self.hello = '안녕하세요'
		self.name = name
		self.age = age
		self.address = address
		self.__wallet = wallet # -> 비공개 속성(__) 사용하기
	# def greeting(self):
	# 	print(f"{self.hello} I'm {self.name}입니다.")
	def pay(self,amount):
		self.__wallet -= amount
		print(f"{self.__wallet}원 남았습니다")

maria = Person("maria", 20,"서울시 서초구 반포동", 10000)
# maria.greeting()
# maria.__wallet() --> 이렇게 할 시 비공개 속성 실행불가
maria.pay(3000)
```

```python
#34.5 연습문제 : 게임 캐릭터 클래스 만들기
class Knight:
	def __init__(self,health, mana, armor):
		self.health = health
		self.mana = mana
		self.armor = armor
	def slash(self):
		print('베기')

x = Knight(health=542.4, mana = 210.3, armor= 38)
print(x.health, x.mana, x.armor)
x.slash()
```

```python
#34.6 심사문제: 게임 캐릭터 클래스 만들기

class Annie:
	def __init__(self,health, mana, AP):
		self.health = health
		self.mana = mana
		self.AP = AP

	def tibbers(self):
		damage = self.AP * 0.65 +400
		print(f"티버: 피해량 {damage:.1f}")

health, mana, AP = map(float,[511.68, 334, 298])
x = Annie(health=health, mana=mana, AP=AP)
x.tibbers()
```
