```python
#break로 반복문끝내기
i = 0
while True:
	print(i)
	i += 20
	if i == 100:
		break
```

```python
#for에서 break로 반복문 끝내기
for i in range(100):
	print(i)
	if i == 10:
		break
```

```python
#continue 코드 실행 건너뛰기
for i in range(10):
	if i % 2 == 0:
		continue
	print(i)
```

```python
#while 반복문에서 continue로 건너뛰기
i = 0
while i < 10:
	i += 1
	if i % 2 ==0:
		continue
	print(i)
```

```python
#입력한 횟수대로 반복하기
count = int(input("숫자"))

i = 0
while True:
	print(i)
	i += 1
	if i == count:
		break
	
```

```python
#Fizz Buzz
#1부터 100까지 숫자 출력하기
# for i in range(10):
# 	print(i)

#3의 배수는 Fizz, 5의 배수는 Buzz
for i in range(100):
	if i % 3 ==0:
		print("Fizz")
	elif i % 5 ==0:
		print("Buzz")
	else:
		print(i)
```

```python
#리스트의 가장 작은 수, 가장 큰 수, 합계 구하기
a = [31, 28, 50, 64, 89]
print(min(a))
print(max(a))
print(sum(a))

smallest = a[0]
for i in a:
	if i < smallest:
		smallest = i
print(smallest)

largest = a[0]
for i in a:
	if i > largest:
		largest = i
print(largest)
```

```python
#리스트 표현식 사용하기 = 리스트 컴프리헨션
a = [i for i in range(10)]
print(a)
b = list(i for i in range(10))
print(b)
c= [i+5 for i in range(10)]
print(c)
d = [j*2 for j in range(10)]
print(d)
```

```python
# 리스트 표현에서 if 조건문 사용하기
## 1. [식 for 변수 in 리스트 if 조건문]
## 2. list(식 for 변수 in 리스트 if 조건문)

a = [i for i in range(10) if i % 2 ==0]
print(a)
```

```python
b = [a + 5 for a in range(10) if a%2 == 1]
print(b)
```

```python
#for 반복문과 if 조건문을 여러번 사용하기
# [식 for 변수1 in 리스트1 if 조건식1 
#  for 변수2 in 리스트2 if 조건식2]
a = [i*j for i in range(2,10) for j in range(1, 10)]
print(a)
```
