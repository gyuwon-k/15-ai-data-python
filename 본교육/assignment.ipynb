```python
# 문제 1. 
# 두 수 사이의 숫자 중  3과 7과 끝나지 않는 숫자 출력하기 

# 입력값 
# 3, 50

# 첫 번째 정수와 두 번째 정수 사이의 숫자 중 3 또는 7로 끝나지 않늣 숫자가 출력되게 만드세요 

a,b = map(int,input("정수 두 개를 입력하시오 : ").split())
if 0<=a<=b:
	num = [i for i in range(a,b+1) if not i%10 ==3 if not i%10 ==7]
	print(num)
else:
	print("잘못된 숫자입니다")
```

```python
# 문제. 1

# 가장 높은 점수, 낮은 점수, 총점, 평균, 분산을 구하는 프로그램을 만들어주세요 

# 입력예시 
# 70 80 85 38 43 30
# 또는 
# 100 18 38 94 93 12 39

# 공백으로 구분되어 몇개가 될지 모름

# 요구사항 
# 1. 함수이름은 스스로 지정해 주세요 
# 2. 가변 인수를 사용해 주세요 

user_input = (map(int,input().split()))

def calculator(*args):
	max_score = max(args)
	min_score = min(args)
	total_score = sum(args)
	avg_score = total_score / len(args)

	squared_deviations = [(x - avg_score) ** 2 for x in args]
	var_score = sum(squared_deviations) / len(args)
	return [max_score, min_score, total_score, avg_score, var_score]

a, b, c, d, e = calculator(*user_input)
print('가장 높은 점수: {0}, 가장 낮은 점수: {1}, 총점: {2}, 평균:{3:.2f}, 분산:{4:.2f}'.format(a,b,c,d,e))
```

```python
# 문제. 2

# prices = {
#     "아메리카노": 2500,
#     "카페라떼": 3500,
#     "바닐라라떼": 4000,
#     "초코케이크": 5500,
#     "샌드위치": 6000
# }

# 함수를 호출하면
# order(아메리카노=2, 카페라떼=1)
# order(바닐라라떼=2, 초코케이크=1, 샌드위치=1)
# order(아메리카노=1, 샌드위치=2)

# 실행결과
# 아메리카노 2개: 5000원
# 카페라떼 1개: 3500원
# 총 결제 금액: 8500원
# --------------------
# 바닐라라떼 2개: 8000원
# 초코케이크 1개: 5500원
# 샌드위치 1개: 6000원
# 총 결제 금액: 19500원
# --------------------
# 아메리카노 1개: 2500원
# 등록되지 않은 상품입니다: 허니브레드
# 총 결제 금액: 2500원
# --------------------

# 요구사항 
# 1. 함수 이름은 order로 만드세요.
# 2. 함수는 **kwargs를 사용해야 합니다.
# 3. kwargs의 key는 상품명, value는 주문 수량입니다.

prices = {
    "아메리카노": 2500,
    "카페라떼": 3500,
    "바닐라라떼": 4000,
    "초코케이크": 5500,
    "샌드위치": 6000
}

def order(**kwargs):
    total = 0
    
    for menu, count in kwargs.items():
        
        if menu in prices:
            price = prices[menu] * count
            total += price
            
            print(f"{menu} {count}개: {price}원")

        else:
            print(f"등록되지 않은 상품입니다: {menu}")
    
    print(f"총 결제 금액: {total}원")
    print("--------------------")


order(아메리카노=2, 카페라떼=1)

order(바닐라라떼=2, 초코케이크=1, 샌드위치=1)

order(아메리카노=1, 허니브레드=1)
```

```python
# 문제 1. 재귀 호출로 합계와 계산 과정 출력하기

# 정수 n을 입력받아 n부터 1까지 더한 결과를 출력하는 프로그램을 만드세요.

# 단, 반드시 재귀 호출을 사용해야 하며, 계산 과정도 함께 출력해야 합니다.

# 입력:
# 5

# 출력: 
# 5 + 4 + 3 + 2 + 1 = 15

# 요구사항: 
# 1. 함수 이름은 recursive_sum으로 만드세요.
# 2. 반복문 for, while을 사용하지 마세요.
# 3. 재귀 호출을 사용하세요.
# 4. n == 1일 때 재귀 호출이 끝나도록 하세요.
# 5. 함수는 계산 과정 문자열과 합계를 함께 반환해야 합니다.
# 6. 최종 출력은 계산과정 = 결과 형태로 출력하세요.

n = int(input())
def recursive_sum(n):
	if n == 1:
		print(f"{n} = ", end='')
		return 1
	else:
		print(f"{n} + ", end='')

	return n + recursive_sum(n-1)
print(recursive_sum(n))
```

```python
# 문제 2. filter()와 lambda를 이용한 게시글 삭제 프로그램

# 다음과 같이 게시글 목록이 주어져 있습니다.


# 필터할 게시글 번호를 입력받아, 해당 id를 가진 게시글을 삭제한 뒤 남은 게시글 목록을 출력하는 프로그램을 작성하세요.

# 입력: 
# 3

# 출력: 
# 삭제 후 게시글 목록
# 1 파이썬 공부 김철수
# 2 오늘 할 일 이영희
# 4 파이썬 복습 김철수
# 5 운동 계획 최지수
posts = [
    {"id": 1, "title": "파이썬 공부", "writer": "김철수"},
    {"id": 2, "title": "오늘 할 일", "writer": "이영희"},
    {"id": 3, "title": "장보기 목록", "writer": "박민수"},
    {"id": 4, "title": "파이썬 복습", "writer": "김철수"},
    {"id": 5, "title": "운동 계획", "writer": "최지수"}
]
filter_num = int(input())
result = list(filter(lambda x: x['id'] != filter_num  ,posts))

for post in result:
    print(f"{post['id']} {post['title']} {post['writer']}")
```

```python
# 문제 3. 클래스를 이용한 21명 학생 성적 관리 프로그램

# 다음 21명의 학생 이름과 점수를 이용하여 학생 성적 관리 프로그램을 만드세요.
# 학생 한 명의 정보를 표현하는 Student 클래스를 만들고, 학생 이름을 기준으로 각 학생 객체를 저장하세요.



import random
names = [
    "권석현", "김규원", "김성준", "김태훈", "김형민", "문홍식", "박규민", "박성민", "박진영", "박찬현",
    "박희재", "손성욱", "신재민", "오은진", "원영배","이도영", "이수현", "이원석", "이진욱", "조현우","홍득훈"
]
student_data = {}
for name in names:
    student_data[name] = [random.randint(60, 100) for _ in range(10)]
# print(student_data)

# 위 코드를 먼저 실행하여 student_data 를 생성하세요 
# student_data를 기반으로 Student의 객체를 생성해주세요 
# 객체의 이름은 student_data의 키 값으로 해주세요 
class Student:
    def __init__(self, name, scores):
		# 이름과 점수 저장
        self.name = name
        self.scores = scores

    def total(self):
		# 총점 반환
        return sum(self.scores)

    def average(self):
		# 평균 반환
        avg = self.total()/len(self.scores)
        return avg
        
    def max_score(self):
		# 가장 높은 점수 반환
        return max(self.scores)

    def min_score(self):
        # 가장 낮은 점수 반환
        return min(self.scores)

    def is_pass(self):
        # 평균 70점 이상이면 합격, 아니면 불합격
        if self.average() >=70:
            return "합격"
        else:
            return "불합격"


    def print_info(self):
        print(f"이름 : {self.name}")
        print(f"점수 : {self.scores}")
        print(f"총점 : {self.total()}")
        print(f"평균 : {self.average():.2f}")
        print(f"최고점 : {self.max_score()}")
        print(f"최저점 : {self.min_score()}")
        print(f"결과 : {self.is_pass()}")
        print("-" * 40)


students = {}

for name in names:

    students[name] = Student(name, student_data[name])


for name in names:

    students[name].print_info()

```
