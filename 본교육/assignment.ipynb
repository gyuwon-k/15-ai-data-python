```python
# 문제 1. 
# 두 수 사이의 숫자 중  3과 7과 끝나지 않는 숫자 출력하기 

# 입력값 
# 3, 50

# 첫 번째 정수와 두 번째 정수 사이의 숫자 중 3 또는 7로 끝나지 않늣 숫자가 출력되게 만드세요 

a,b = map(int,input("정수 두 개를 입력하시오 : ").split())
if 0<=a<=b:
	num = [i for i in range(a,b+1) if not i%10 ==3 if not i%10 ==7]
	print(num)
else:
	print("잘못된 숫자입니다")
```

```python
# 문제 3. 
# 입력값을 참조하여 첫 번째 정수 부터 두 번째 정수까지 숫자를 역으로 출력하면서 
# 3의 배수일때는 Fizz, 7의 배수일때는 Buzz, 그외는 숫자를 출력하세요 

# 입력값 
# 100 10

a,b = map(int,input("정수 두 개를 입력하시오 : ").split())
if 0<=b<=a:
	for i in range(a,b,-1):
		if i%3 ==0 and i%7 ==0:
			print("FizzBuzz")
		elif i%3 ==0:
			print("Fizz")
		elif i%7 ==0:
			print("Buzz")
		else:
			print(i)
else:
	print("잘못된 숫자입니다")
```

```python
# 문제 2. 
# 표준 입력으로 역삼각형의 높이가 입력됩니다. 입력된 높이만큼 역삼각형의 "*"을 출력하는 프로그램을 작성하세요 

# 입력값
# 5

# 출력값 
# *********
#  *******
#   *****
#    ***
#     *

# 요구사항:
# 중첩 for를 사용

height = int(input("높이를 입력하시오 : "))

for i in range(height):
	for j in range(i):
		print(" ", end="")
	for k in range(2*(height-i)-1):
		print("*", end="")
	print()

```
