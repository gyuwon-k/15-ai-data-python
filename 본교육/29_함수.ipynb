```python
#위치 인수를 사용하는 함수를 만들고 호출하기
def print_number(a,b,c):
	print(a)
	print(b)
	print(c)

print_number(10,20,30)
```

```python
# 가변 인수 만들기
## def 함수이름(*매개변수):
## 		코드
def print_numbers(*args):
	print(args)
	for arg in args:
		print(arg)
		
```

```python
print_numbers(10)
print_numbers(10, 20, 30, 40)
```

```python
y = [10,20,30,40]
print_numbers(*y)
```

```python
def print_numberss(a, *args):
	print(a)
	print(args)

print_numberss(1)
print_numberss(1, 10, 20)
```

```python
#키워드 인수 사용하기

def personal_info(name, age, address):
	print('이름: ', name)
	print('나이: ', age)
	print('주소: ', address)

personal_info("홍길동", 30, "서울시 용산구")
```

```python
# 키워드 인수를 이용하여 가변 인수 만들기
def personal_info(**kwargs):
	print(kwargs)
	for kw, arg in kwargs.items():
		print(kw, ': ', arg, sep='')

personal_info(name = '홍길동')
```

```python
personal_info(age=30, address="서울시 용산구", name = "홍길동")
```

```python
#30.6 가장 높은 점수를 구하는 함수 만들기
korean, english, math, science = 100, 86, 81, 91

def get_max_score(*args):
	return max(args)

max_score = get_max_score(korean, english, math, science)
print('높은 점수: ',max_score)

max_score = get_max_score(english,science)
print('높은 점수: ',max_score)
```

```python
#30.7 심사문제 : 가장 낮은 점수, 높은 점수와 평균점수를 구하는 함수 만들기
korean, english, math, science = map(int,input().split())

def get_min_max_score(*args):
	return [min(args), max(args)]

def get_avg(**kwargs):
	scores = kwargs.values()
	c = len(kwargs)
	return sum(scores)/c


min_score, max_score = get_min_max_score(korean, english, math, science)
avg_score = get_avg(korean = korean, english = english, math = math, science = science)
print('낮은 점수 : {0:.2f}, 높은 점수: {1:.2f}, 평균 점수 {2:.2f}'.format(min_score, max_score, avg_score))

min_score, max_score = get_min_max_score(english,science)
avg_score = get_avg(english = english, science = science)
print('낮은 점수 : {0:.2f}, 높은 점수: {1:.2f}, 평균 점수 {2:.2f}'.format(min_score, max_score, avg_score))
```

```python
#재귀함수
def hello(count):
	if count ==0:
		return 50
	
	print('hello, world', count)
	count -= 1
	return hello(count)		#자신을 다시 불러와서 종료조건을 만족할 때까지 반복
hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기

def factorial(n):
	if n ==1:
		return 1
	return n * factorial(n-1)   #재귀로인해 5*4*3*2*1 이렇게 조건을 만족하고 종료됨.

factorial(5)
```

```python
#31.4 연습문제: 재귀호출로 회문 판별하기
def is_palindrome(word):
	if len(word) < 2:
		return True
	if word[0] != word[-1]:
		return False
	return is_palindrome(word[1:-1])
```

```python
#31.5 심사문제 : 재귀호출로 피보나치 수 구하기 / n까지의 피보나치 수
#n = 1,2인경우 1 , n = 3인경우부터 시작됨.  f(n) = f(n-1) + f(n-2)
def piv(n):
    if n == 1 or n == 2:
        return 1
    else:
        return piv(n - 1) + piv(n - 2)

n = int(input())

print(piv(n))
```

```python
# 32.4 연습문제 : 이미지 파일만 가져오기

files = ['font', '1.png', '11.gif', '2.jpg', '3.png', 'table.xslx', 'spec.docx']
print(list(filter(lambda x : x.find('.jpg') != -1 or x.find('.png') != -1,files)))
```

#클로저 사용하기

```python
# 함수 안에 전역 변수 변경하기
x = 10				#전역변수
def foo():		
	x = 20			#지역변수
	print(x)

foo()				#지역변수 출격
print(x)			#전역변수 출격


def foog():
	global x
	x = 20
	print(x)

foog()
print(x)
```

```python
# 함수 안에서 함수 만들기
def print_hello():
	hello = 'hello world!'
	def print_msg():
		print(hello)
	print_msg()

print_hello()
```

```python
#지역 변수의 범위
def A():
	x = 10
	def B ():
		x = 20
	B()
	print(x)
A()
```

```python
# nonlocal이 변수를 찾는 순서

def A():
	x = 10
	y = 100
	def B():
		x= 20
		def C():
			nonlocal x
			nonlocal y
			x = x+30
			y = y + 300
			print(x)
			print(y)
		C()
	B()

A()
```

```python
# global로 전역 변수 사용하기
x = 1
def A():
	x = 10
	def B():
		x = 20
		def C():
			global x
			x += 10
			print(x)
		C()
	B()
A()
			
```

```python
#lambda로 클로저만들기

def calc():
	a = 3
	b = 5
	return lambda x: a*x+b

c = calc()
print(c(1),c(2),c(3))
```
