```python
#23.7
col, row = map(int,input("행과 열을 차례대로 기입하시오 : ").split())
matrix = []
#지뢰 만들기 *은 지뢰, .은 지뢰가 없는 곳
for i in range(row):
	matrix.append(list(input()))

result = matrix.copy()
#인접한 것 리스트화
#지뢰의 위치를 우선적으로 찾고 해당 위치로부터 떨어진 위치 계산
# *이 있으면 continue로 넘어가고 .이 있으면 주변에 *이 몇 개 있는지 계산해서 그 수를 넣기.
#카운트하는 기준은 8방향으로 1칸
for i in range(row):
	for j in range(col):
		if matrix[i][j] == "*":
			continue
		else:
			count = 0
			for row_c in range(i-1,i+2):			#현재 행 기준 위아래
				for col_c in range(j-1,j+2):		#현재 열 기준 양옆
					if 0 <= row_c < row and 0 <= col_c < col:		
						if matrix[row_c][col_c] == "*":
							count += 1
		result[i][j] = count

for i in range(row):
    print(''.join(map(str, result[i])))
```

```python
# 딕셔너리 표현식에서 if 조건문 사용하기
x = {
	"a": 10,
	"b": 20,
	"c": 30,
	"d": 40
}

# for key, value in x.items():  #이렇게 하면 오류가 발생. 루프가 도는 와중에 딕셔너리가 변경되기 때문. 딕셔너리는 index(순서)가 아닌 key값을 찾는 개념이기에 오류가 발생함.
for key, value in list(x.items()): # list를 사용하면 복제본에서 삭제하는 것. 기본 딕셔너리x는 변경되지 않음
	if value == 20:
		del x[key]
print(x)
```
