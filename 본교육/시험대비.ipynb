```python
# 문제 1



# 출력결과 
# ===== 학생 성적 결과 =====

# 이름: 김철수

# 점수: [90, 80, 70]

# 평균: 80.0



# 이름: 이영희

# 점수: [100, 95, 90]

# 평균: 95.0



# 이름: 박민수

# 점수: [60, 70, 65]

# 평균: 65.0

# ===== 전체 결과 =====

# 전체 학생 수: 3명

# 전체 평균: 80.0
students = {

    "김철수": [90, 80, 70],
    
    "이영희": [100, 95, 90],
    
    "박민수": [60, 70, 65]

}

print("===== 학생 성적 결과 =====")
total = 0
for name, score in students.items():
	avg = sum(score)/len(score)
	print(f"이름: {name}\n")
	print(f"점수: {score}\n")
	print(f"평균: {avg}\n")
	total += sum(score) 

value_list = list(students.values())
subject_count = len(value_list[0])*len(students)
print("===== 전체결과 =====")
print(f"전체 학생 수 : {len(students)}")
print(f"전체 평균: {total/subject_count}")
```

```python
# 문제 1. 
# 표준 입력으로 2차원 리스트의 가로(col)와 세로(row)가 입력되고 그 다음 줄부터 리스트의 요소로 들아갈 문자가 입력됩니다. 
# 이때 2차원 리스트 안세어 *는 지뢰이고, .은 지뢰가 압니다. 지뢰가 아닌 요소에는 인접한 지뢰의 개수를 출력하는 프로그램을 만드세요 



# 입력 
# 3 3
# .**
# *..
# .*.

# 출력
# 2**
# *43
# 2*1

import copy
col, row = map(int,input().split())

matrix = []

for i in range(row): 
    matrix.append(list(input()))

result = copy.deepcopy(matrix)
for i in range(row):
    for j in range(col):
        if matrix[i][j] == '*':
            continue
        else:
            count = 0
            for row_c in range(i-1,i+2):
                for col_c in range(j-1, j+2):
                       if 0 <= row_c < row and 0 <= col_c < col:		
                              if matrix[row_c][col_c] == "*":
                                      count += 1
            result[i][j] = count
                
for i in range(row):
       print(''.join(map(str,result[i])))
         
```
