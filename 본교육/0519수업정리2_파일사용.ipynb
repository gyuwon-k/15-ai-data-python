#파일 사용하기

```python
#파일에 문자열 쓰기
## 파일객체 = open(파일이름, 파일모드)
## 파일객체.write(문자열)

file = open("./files/hello.txt", "w")
file.write("Hello, World!\n")
file.write("Welcome to Python programming.\n")
file.close()
```

```python
#자동으로 파일 객체 닫기
## with open(파일이름, 파일모드) as 파일객체:
##      코드

with open("./files/hello.txt", "r") as file:
	s= file.read()
	print(s)
```

```python
#반복문으로 문자열 여러 줄을 파일에 쓰기

with open("./files/hello3.txt", "w") as file:
	for i in range(3):
		file.write(f"hello world {i}\n")
```

```python
# 리스트에 들어있는 문자열을 파일에 쓰기
lines = ["안녕하세요\n", "파이썬\n", "코딩도장입니다\n"]

with open("./files/hello4.txt", "w", encoding="utf-8") as file:
	file.writelines(lines)
```

```python
# 파일의 내용을 한 줄씩 리스트로 가져오기
## 변수 = 파일객체.readlines()
with open("./files/hello4.txt", "r", encoding="utf-8") as file:
	lines = file.readlines()
	print(lines)
```

```python
#for 반복문으로 파일의 내용을 줄 단위로 읽기
with open("./files/hello4.txt", "r", encoding="utf-8") as file:
	for line in file:
		print(line, end="")
```

```python
#파이썬 객체를 파일에 저장하기, 가져오기

import pickle
name = "james"
age = 17
address = "서울시 서초구"
scores = {
	"korean": 90,
	"english": 85,
	"math": 95,
	"science": 80
}

with open("./files/james.p", "wb") as file:
	pickle.dump((name, age, address, scores), file)
```

```python
import pickle
with open("./files/james.p", "rb") as file:
	name, age, address, scores = pickle.load(file)
	print(name)
	print(age)
	print(address)
	print(scores)
```

```python
#27.5 파일에서 10자 이하인 단어 개수 세기
with open("./files/words.txt", "r") as file:
	words = file.read().split()
	count = 0
	for word in words:
		if len(word) <= 10:
			count += 1
	print(count)
```

```python
#27.6 심사문제: 특정 문자가 들어있는 단어 찾기 --문자 c가 포함된 단어를 각 줄에 출력하는 프로그램
with open("./files/words.txt", "r") as file:
	words = file.read().split()
	for word in words:
		if "c" in word:
			print(word)
```

#회문 판별과 N-gram 만들기

```python
#회문 판별하기
# 반복문으로 문자 검사하기

word = input("문자열을 입력하세요: ")
is_palindrome = True
for i in range(len(word) // 2):
	if word[i] != word[-(i + 1)]:
		is_palindrome = False
		break
print(is_palindrome)

```

```python
#시퀀스 뒤집기로 문자 검사하기
word = input("단어를 입력하세요 : ")

print(word == word[ : : -1])
```

```python
# 리스트와  reversed 사용하기

word = "level"

list(word) == list(reversed(word))
```

```python
word = 'level'

word == "".join(reversed(word))
```

```python
# N-gram 만들기
# 반복문으로 N-gram 출력하기

text = "hello"
for i in range(len(text)-1):
	print(text[i], text[i+1], sep='')
```

```python
text = "This is python script"
words = text.split()

for i in range(len(words) -1):
	print(words[i], words[i+1])
```

```python
#zip으로 2-gram만들기
text ="hello"

two_gram = zip(text, text[1:])

for i in two_gram:
	print(i[0], i[1], sep = '')
```
