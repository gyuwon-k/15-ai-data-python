```python

scores = input("국어, 영어, 수학, 과학 점수를 입력하시오 : ").split()
score = list(map(int,scores))
avg = (sum(score)/len(score))
if min(score)<0 or max(score)>100 :
	print('잘못된 점수')
elif avg>=80:
	print('합격')
else:
	print('불합격')
```

```python
#15.4
age = int(input("나이를 입력하시오: "))
card = 9000
if 7<=age<=12:
	card -= 650
	print(f"잔액 : {card}")
elif 13<=age<=18 :
	card -= 1050
	print(f"잔액 : {card}")
else:
	card -= 1250
	print(f"잔액 : {card}")
```

```python
# while 조건문
#		반복할 코드
i = 0
while i < 100:
	print("hello world")
	i += 20
```
