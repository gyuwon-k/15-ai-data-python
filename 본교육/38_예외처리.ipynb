```python
#  try except

# try :
# 	실행할 코드
# except:
# 	예외가 발생했을 때 처리하는 코드

try:
	x= int(input())
	y= 10 / x
	print(y)
except:
	print("예외가 발생했습니다")
```

```python
# 특정 예외만 처리하기
# try:
# 	실행할 코드
# except 예외이름:
# 	예외가 발생했을 때 처리하는 코드

y = [10, 20, 30]

try:
	index, x = map(int,input().split())
	print(y[index] / x)
except ZeroDivisionError:
	print("숫자를 0으로 나눌 수 없습니다")
except IndexError:
	print("잘못된 인덱스")
except ValueError:
	print("잘못된 입력")
```

```python
# #예외의 에러 메세지 받아오기
# try:
# 	실행할 코드
# except 예외 as 변수:
# 	예외가 발생했을 때 처리하는 코드

y = [10 ,20 , 30]

try:
	index, x = map(int,input().split())
	print(y[index] / x)
except ZeroDivisionError as e:
	print("숫자를 0으로 나눌 수 없습니다",e)
except IndexError as e:
	print("잘못된 인덱스",e)	
```

```python
# else와 finally 사용하기
# try:
# 	실행할 코드
# except:
# 	예외가 발생했을 때 처리하는 코드
# else:
# 	예외가 발생하지 않았을 때 실행할 코드
# finally:
	# 예외 발생 여부와 상관없이 항상 실행할 코드

try:
	x = int(input())
	y = 10 / x
except ZeroDivisionError:
	print("숫자를 0으로 나눌 수 없습니다")
else:
	print(y)
finally:
	print('코드 실행이 끝났습니다')

```

```python
def three_multiple():
    try:
        x = int(input('3의 배수를 입력하세요: '))
        if x % 3 != 0:                                 # x가 3의 배수가 아니면
            raise Exception('3의 배수가 아닙니다.')    # 예외를 발생시킴
        print(x)
    except Exception as e:                             # 함수 안에서 예외를 처리함
        print('three_multiple 함수에서 예외가 발생했습니다.', e)
        raise    # raise로 현재 예외를 다시 발생시켜서 상위 코드 블록으로 넘김
 
try:
    three_multiple()
except Exception as e:                                 # 하위 코드 블록에서 예외가 발생해도 실행됨
    print('스크립트 파일에서 예외가 발생했습니다.', e)
```
