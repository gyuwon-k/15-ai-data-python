```python
# 문자열 판단하기

import re
re.match('hello', 'hello, world!')
```

```python
# 문자열이 맨 앞에 오는지 맨 뒤에 오는지 판단하기
# re.search("패턴", "문자열")
print(re.search('^Hello', 'Hello, world!'))     
print(re.search('world!$', 'Hello, world!'))
```

```python
# 범위 판단하기
# [0-9]*
# [0-9]+

print(re.match('[0-9]*', '1234'))
print(re.match('[0-9]', '1234'))
print(re.match('[0-9]+', '1234'))
print(re.match('[0-9]+', 'abcd'))
```

```python
# 문자?
# [0-9]?
print(re.match('abc?d', 'abd'))
print(re.match('abc?d', 'abcd'))
print(re.match('ab[0-9]?c', 'ab3c') )
print(re.match('ab[0-9]?c', 'abc') )
```

```python
# 문자 개수 판단하기
## 문자(개수)
## (문자열)(개수)

print(re.match('h{3}', 'hhhello'))
print(re.match('h{3}', 'hhello'))
```

```python
## [0-9](개수)
print(re.match("[0-9]{3}-[0-9]{4}-[0-9]{4}", "010-2222-3333"))
```

```python
## (문자){시작개수, 끝개수}
## (문자열){시작개수, 끝개수}
## [0-9]{시작개수, 끝개수}

print(re.match('[0-9]{2,3}-[0-9]{3,4}-[0-9]{4}', '02-100-1000'))
```

```python
## a-z
## A-Z
print(re.match('[a-zA-Z0-9]+', 'Hello1234'))
print(re.match('[a-zA-Z0-9]+', 'Hello1234'))
print(re.match('[A-Z0-9]+', 'hello'))
```

```python
# 가 - 힇
print(re.match("[가-힣]+", "홍길동"))
print(re.match("[가-힣]+", "ABC"))
```

```python
# 특정 문자 범위에 포함되지 않는지 판단하기

# [^범위]*
# [^범위]+

print(re.match('[^A-Z]+', 'Hello'))    # 대문자를 제외. 대문자가 있으므로 패턴에 매칭되지 않음
```

```python
# ^[범위]*
# ^[범위]+
re.search('^[A-Z]+', 'Hello')        # 대문자로 시작하므로 패턴에 매칭됨
```

```python
# [범위]*$
re.search('[0-9]+$', 'Hello1234')    # 숫자로 끝나므로 패턴에 매칭됨
```

```python
# 특수 문자 판단하기
re.search('\*+', '1 ** 2')                    # *이 들어있는지 판단
# \\d: [0-9]와 같음. 모든 숫자
# \\D: [^0-9]와 같음. 숫자를 제외한 모든 문자
# \\w: [a-zA-Z0-9_]와 같음. 영문 대소문자, 숫자, 밑줄 문자
# \\W: [^a-zA-Z0-9_]와 같음. 영문 대소문자, 숫자, 밑줄 문자를 제외한 모든 문자
print(re.match('\d+', '1234'))          # 모든 숫자이므로 패턴에 매칭됨
print(re.match('\D+', 'Hello'))
print(re.match('\w+', 'Hello_1234'))
print(re.match('\W+', '(:)'))
```

```python
# 공백 처리하기
# \\s: [ \t\n\r\f\v]와 같음. 공백(스페이스), \t(탭) \n(새 줄, 라인 피드), \r(캐리지 리턴), \f(폼피드), \v(수직 탭)을 포함
# \\S: [^ \t\n\r\f\v]와 같음. 공백을 제외하고 \t, \n, \r, \f, \v만 포함
print(re.match('[a-zA-Z0-9 ]+', 'Hello 1234'))     # ' '로 공백 표현
print(re.match('[a-zA-Z0-9\s]+', 'Hello 1234'))    # \s로 공백 표현

```

```python
# 그룹 사용하기
m = re.match('([0-9]+) ([0-9]+)', '10 295')
print(m.group(1))
```

```python
# 패턴에 매칭되는 모든 문자열 가져오기
re.findall('[0-9]+', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
# 문자열 바꾸기
print(re.sub('apple|orange', 'fruit', 'apple box orange tree'))    # apple 또는 orange를 fruit로 바꿈
print(re.sub('[0-9]+', 'n', '1 2 Fizz 4 Buzz Fizz 7 8'))    # 숫자만 찾아서 n으로 바꿈
```

```python
# 교체함수(매치객체)
# re.sub('패턴', 교체함수, '문자열', 바꿀횟수)
def multiple10(m):        # 매개변수로 매치 객체를 받음
     n = int(m.group())    # 매칭된 문자열을 가져와서 정수로 변환
     return str(n * 10)    # 숫자에 10을 곱한 뒤 문자열로 변환해서 반환

re.sub('[0-9]+', multiple10, '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
# 찾은 문자열을 결과에 다시 사용하기
print(re.sub('([a-z]+) ([0-9]+)', '\\2 \\1 \\2 \\1', 'hello 1234')    # 그룹 2, 1, 2, 1 순으로 바꿈
)
```
